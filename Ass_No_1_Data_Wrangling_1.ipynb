# Deal with missing data
# How to deal with missing data?

1) drop data
*   drop the whole row
*   drop the whole column

2) replace data
*   replace it by mean
*   replace it by frequency
*   replace it based on other functions

Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely. We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns

```python
# Find avg normalized-losses column
avg_norm_loss = df['normalized-losses'].astype('float').mean()
df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace=True)
df['normalized-losses'].isnull().sum()
```

```python
avg_stroke = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
df['stroke'].isnull().sum()
```

```python
avg_compression_ratio = df['compression-ratio'].astype('float').mean()
df['compression-ratio'].replace(np.nan, avg_compression_ratio, inplace=True)
df['compression-ratio'].isnull().sum()
```

```python
avg_city_mpg = df['city-mpg'].astype('float').mean()
df['city-mpg'].replace(np.nan, avg_city_mpg , inplace=True)
df['city-mpg'].isnull().sum()
```

```python
avg_highway_mpg = df['highway-mpg'].astype('float').mean()
df['highway-mpg'].replace(np.nan, avg_highway_mpg, inplace=True)
df['highway-mpg'].isnull().sum()
```

```python
# num of door - Replace na values of it with its frequency , 
# here four has occured multiple time so replace it with 4

avg_num_of_doors = df['num-of-doors'].value_counts().idxmax()
df['num-of-doors'].replace(np.nan, avg_num_of_doors, inplace=True)
df['num-of-doors'].isnull().sum()
```

```python
# drop unnamed column - rename them & then drop them
df.rename({"Unnamed: 24":"a"}, axis="columns", inplace=True)
df.drop(["a"], axis=1, inplace=True)

df.rename({"Unnamed: 25":"b"}, axis="columns", inplace=True)
df.drop(["b"], axis=1, inplace=True)

df.isnull().sum()
```

```python
# Correct Data Format
# dtype - for check data type
# astype - for changing data type
print("Before Datatype conversion")
df.dtypes
```

```python
df[['normalized-losses', 'city-mpg', 'highway-mpg', 'price']] = df[['normalized-losses', 'city-mpg','highway-mpg', 'price']].astype('int')
df[['stroke', 'compression-ratio']] = df[['stroke', 'compression-ratio']].astype('float')
```

```python
print("After Datatype conversion")
df.dtypes
```
