```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Iris.csv')
df
```

```python
import sklearn
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
df.SepalLengthCm = number.fit_transform(df.SepalLengthCm)
df.SepalWidthCm = number.fit_transform(df.SepalWidthCm)
df.PetalLengthCm = number.fit_transform(df.PetalLengthCm)
df.PetalWidthCm = number.fit_transform(df.PetalWidthCm)
df.Species = number.fit_transform(df.Species)

df
```

```python
features = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm' , 'PetalWidthCm']]
target = df['Species']
```

```python
model = GaussianNB()
model.fit(features,target)
# scenario for prediction: outlook = rainy(1),temp = cool(0), Humidity = high(0), wind= strong(0)
scenario = [[1,0,0,0]]
print("Model prediction",model.predict(scenario))
print("Probability prediction",model.predict_proba(scenario))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

print("\nConfusion Matrix\n", confusion_matrix(target,model.predict(features)))
print("\nClassification report\n", classification_report(target,model.predict(features)))
print("\nAccuracy score\n", accuracy_score(target,model.predict(features)))
```
